```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
file_to_load = "C:/Users/bvkka/Desktop/states_all.csv"

# Read US Education file and store into Pandas data frame
df = pd.read_csv(file_to_load)

df.head()
```

```python
df.columns
```

```python
# data frame with Selected columns
df[['PRIMARY_KEY', 'STATE', 'YEAR','TOTAL_EXPENDITURE','GRADES_PK_G', 'GRADES_KG_G','GRADES_4_G', 'GRADES_8_G', 'GRADES_12_G', 'GRADES_1_8_G','GRADES_9_12_G']].head(5)
```

```python
df.STATE.unique()
```

```python
#Checking NA's in each column 
#count_nan = len(df) - df.count()
#count_nan
df.isnull().sum(axis = 0)
```

```python
#Removing NA's from the selected columns 
df1 =df[['PRIMARY_KEY','STATE','TOTAL_EXPENDITURE','YEAR','GRADES_ALL_G']].dropna(axis=0,how='any')

#df1.isnull().sum(axis = 0)
count_nan = len(df1) - df1.count()
count_nan
```

```python
#Removing Duplicates from Selected columns
df2=df1.drop_duplicates(subset=['PRIMARY_KEY'],keep=False).copy()
```

```python
#Creating a new column 'Expenditure per student ' and then adding it to the data frame
df2['EXP_PER_STUDENT'] = df2['TOTAL_EXPENDITURE']/df2['GRADES_ALL_G']

df2.head()
```

```python
#Set the DataFrame index to'STATE' column.

df3=df2.set_index('STATE')
df3.head()

```

```python
#Selections of columns using the loc method are based on the index of the data frame 
#Where the index is set on a DataFrame
#the .loc method directly selects based on index values of any rows.

df4=df3.loc[:,['EXP_PER_STUDENT','YEAR']]
df4.head()
```

```python
#Grouping by YEAR and then finding the top 5 and bottom 5 states based on Expenditure Per Student.
df5 = df4.groupby('YEAR')['EXP_PER_STUDENT'].nlargest(5)
df6 = df4.groupby('YEAR')['EXP_PER_STUDENT'].nsmallest(5)
df5.head()
```

```python
#Reset the index of the DataFrame, and use the default one instead. Basically, we are reseting the Year Index to default.
df7=df5.to_frame().reset_index()  #for top 5
df8=df6.to_frame().reset_index()  #for bottom 5
df7.head()
```

```python
#Merging 2 data frames using inner join on 2 columns

#for top5
df9=pd.merge(df7,df2.loc[:,['YEAR','STATE','GRADES_ALL_G','TOTAL_EXPENDITURE']],how='inner',on=['YEAR','STATE'])

#for bottom 5 
df10 = pd.merge(df8,df2.loc[:,['YEAR','STATE','GRADES_ALL_G','TOTAL_EXPENDITURE']],how='inner',on=['YEAR','STATE'])


df9.head() 
```

```python
# And I need to transform my categorical column (state) in a numerical value state1->1, state2->2...
df9['STATE']=pd.Categorical(df9['STATE'])
df10['STATE']=pd.Categorical(df10['STATE'])

df9['STATE'].head()
```

```python
#Data Visualization


# libraries
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
% matplotlib notebook
from matplotlib import animation
from matplotlib import cm   #color maps  [https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html]
import matplotlib.gridspec as gridspec






tmp=df9

#function for animation 
#we are creating a frame, there will be many frames from year 1992 to 2017 (25 in total)
def update(curr):
        
 
        #for animation
        if curr == len(df9.YEAR.unique()): 
            a.event_source.stop()  #for animation to stop when it exceeds the last year
        ax.cla() #to clear previous frames -> or else we will see many frames on top of each other
        bx.cla()
       
        
        #color map
        cmap = plt.cm.Reds
       
        nor = matplotlib.colors.Normalize(vmin=tmp[tmp['YEAR']==df9.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT')['STATE'].
                                          cat.codes.min()
                                            , vmax=tmp[tmp['YEAR']==df9.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT')
                                          ['STATE'].cat.codes.max())
        #values for colors normalizing 0-10
        #vmin and vmax (0 -> 10) 
                
        
        
        
        #Expenditure per student value  (for x axis) -> we get an array here
        a = tmp[tmp['YEAR']==df9.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT')['EXP_PER_STUDENT'].values
        
        #Total Expenditure value  (for y axis)
        b = tmp[tmp['YEAR']==df9.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT')['TOTAL_EXPENDITURE'].values
        
        #Category values for states  (for bubble color)
        d = tmp[tmp['YEAR']==df9.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT')['STATE'].cat.codes.values
        
        #Total Number of students per 2000 (for the size of the bubble)
        e = tmp[tmp['YEAR']==df9.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT')['GRADES_ALL_G']/2000
        
        
        
                #scatter plot
                #DataFrame.plot.scatter(x, y, s=None, c=None, **kwds)[source]
                #Create a scatter plot with varying marker point size and color
                
        ax.scatter(a,b,s=e,c=cmap(nor(d)), alpha=0.6,edgecolors="white")
           # label  for scatter plot   
        ax.set_title("Year: "+str(df9.YEAR.unique()[curr]) )
        ax.set_xlabel("Expenditure Per Student")
        ax.set_ylabel("Total Expenditure")
        ax.set_xlim(0, 30)       
        ax.set_ylim(0, 7e7)
        
        
        #We are creating a legend (another plot )
          
        #c = tmp[tmp['YEAR']==df9.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT').sort_values('GRADES_ALL_G')['STATE'].values
        ##Another way using zip function -> it creates a tuple of all lists . Zipping multiple lists into one list of tuple (Key, value)
        #[https://campus.datacamp.com/courses/pandas-foundations/data-ingestion-inspection?ex=6]
        #You can use the lists to create lists of tuples and create a dictionary from it. Then, this dictionary can be used to construct a dataframe.

        #zip() function creates the objects and that can be used to produce single item at a time. This function can create pandans DataFrames by merging two lists.
        
        for c,d,e in zip( tmp[tmp['YEAR']==df9.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT').sort_values('GRADES_ALL_G')['STATE'].values,
                                  tmp[tmp['YEAR']==df9.YEAR.unique()[curr]].sort_values('GRADES_ALL_G')['STATE'].cat.codes.values,
                                    (tmp[tmp['YEAR']==df9.YEAR.unique()[curr]].sort_values('GRADES_ALL_G')['GRADES_ALL_G']/2000).values):
            
            
                bx.scatter( [],[],s=e,c=cmap(nor(d)),label=c, alpha=0.6,edgecolors="white")
                bx.legend(title= 'States',   labelspacing = 3)
                
       
                bx.axis('off')
      
fig=plt.figure(figsize=(10, 6) )




gspec = gridspec.GridSpec(1, 3)   # grid spec with 1 row and 3 columns 
ax=plt.subplot(gspec[:2])

bx=plt.subplot(gspec[2])
a = animation.FuncAnimation(fig, update, interval=2000)



```
