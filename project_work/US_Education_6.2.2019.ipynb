```python
dfstate_year =df[['PRIMARY_KEY','STATE','TOTAL_EXPENDITURE','YEAR','GRADES_ALL_G']].dropna(axis=0,how='any')
```

```python
dfstate_year=dfstate_year.drop_duplicates(subset=['PRIMARY_KEY'],keep=False)
```

```python
dfstate_year['EXP_PER_STUDENT'] = dfstate_year['TOTAL_EXPENDITURE']/dfstate_year['GRADES_ALL_G']
df.head()
```

```python
dfstate_year
```

```python

df4=dfstate_year
df5=df4.set_index('STATE').loc[:,['EXP_PER_STUDENT','YEAR']].groupby('YEAR')['EXP_PER_STUDENT'].nsmallest(5)

df5=df5.to_frame().reset_index()
df5=pd.merge(df5,dfstate_year.loc[:,['YEAR','STATE','GRADES_ALL_G','TOTAL_EXPENDITURE']],how='inner',on=['YEAR','STATE'])
df5
```

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
my_dpi=96


import pandas as pd

```

```python
# And I need to transform my categorical column (continent) in a numerical value group1->1, group2->2...
df5['STATE']=pd.Categorical(df5['STATE'])

df5['STATE'].head()
```

```python
#For each year:
% matplotlib notebook

from matplotlib import animation
import matplotlib
from matplotlib import cm   #color maps
import matplotlib.gridspec as gridspec
tmp=df5
def update(curr):
        
 #initialize a figure
        #for animation
        if curr == len(df5.YEAR.unique()): 
            a.event_source.stop()  #for animation to stop
        ax.cla() #to clear old data / stop aggregating data
        bx.cla()
       
        
        #color map
        cmap = plt.cm.Reds
       
        
        #values for colors normalizing 0-10
        nor = matplotlib.colors.Normalize(vmin=tmp[tmp['YEAR']==df5.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT')['STATE'].
                                          cat.codes.min()
                                            , vmax=tmp[tmp['YEAR']==df5.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT')
                                          ['STATE'].cat.codes.max())
        
        
        
        for a,b,d,e in zip(tmp[tmp['YEAR']==df5.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT')['EXP_PER_STUDENT'].values,
                       tmp[tmp['YEAR']==df5.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT')['TOTAL_EXPENDITURE'].values,
                               
                                  tmp[tmp['YEAR']==df5.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT')['STATE'].cat.codes.values,
                                 (tmp[tmp['YEAR']==df5.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT')['GRADES_ALL_G']/2000).values):
               
                #scatter plot
                ax.scatter(a,b,s=e,c=cmap(nor(d)), alpha=0.6,edgecolors="white")
                
         # label     
        ax.set_title("Year: "+str(df5.YEAR.unique()[curr]) )
        ax.set_xlabel("Expenditure Per Student")
        ax.set_ylabel("Total Expenditure")

                    
        for c,d,e in zip( tmp[tmp['YEAR']==df5.YEAR.unique()[curr]].sort_values('EXP_PER_STUDENT').sort_values('GRADES_ALL_G')['STATE'].values,
                                  tmp[tmp['YEAR']==df5.YEAR.unique()[curr]].sort_values('GRADES_ALL_G')['STATE'].cat.codes.values,
                                    (tmp[tmp['YEAR']==df5.YEAR.unique()[curr]].sort_values('GRADES_ALL_G')['GRADES_ALL_G']/2000).values):
            
            
                bx.scatter([],[],s=e,c=cmap(nor(d)),label=c, alpha=0.6,edgecolors="white")
        bx.legend(title= 'States',   labelspacing = 3)
       
        bx.axis('off')
      

fig=plt.figure(figsize=(10, 6) )



gspec = gridspec.GridSpec(1, 3)
ax=plt.subplot(gspec[:2])
bx=plt.subplot(gspec[2])
a = animation.FuncAnimation(fig, update, interval=1000)



```
